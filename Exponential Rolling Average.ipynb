# Comparison of Exponent Average & Sample Average

### Sample Average

M(T) = (x_1 + ... x_T)/T = (1-1/T) * M(T - 1) + (1/T) * X_T

As sample size increases, the impact of X_T will decrease becuase 1/T is decreasing

### Exponential Average

let (1 - 1/T) be fixed

M(T) = decay_rate * M(T-1) + (1-decay_rate) * X_T

The impact of X_T will be larger camparing to sample average

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Construct a data set with x range from 0 to 300 and y = sin(x) + noise

```python
x = np.linspace(0,300,1000)
```

```python
noise = np.random.normal(0,1,1000)
```

```python
y = np.sin(0.03*x) +noise
```

```python
plt.plot(x,y)
plt.show()
```

#### Exponential Mean

```python
decay_rate = 0.99
exp_mean = []
em = 0
for el in y:
    em = decay_rate * em + (1- decay_rate) * el
    exp_mean.append(em)
```

#### Sample Mean

```python
sample_mean = []
sm = 0
for i in range(len(y)):
    sm = np.average(y[0:i])
    sample_mean.append(sm)
```

#### Exponential Mean vs Sample Mean

```python
plt.plot(x,y, label = 'original y')
plt.plot(x,sample_mean, label = 'sample mean', color = 'red')
plt.plot(x,exp_mean, label = 'exponential mean', color = 'orange')
plt.legend()
plt.show()
```

#### Sample mean trend to be stable when sample size getting larger while exponential mean fit to the curve better. This is due to larger impact of latest added point value.
